### Table 4-6. Basic array statistical methods

| Method           | Description                                                                 | Example (Python) |
|------------------|-----------------------------------------------------------------------------|------------------|
| `sum`            | Sum of all the elements in the array or along an axis; zero-length arrays have sum 0 | `np.array([1,2,3]).sum()` → `6` |
| `mean`           | Arithmetic mean; invalid (returns `NaN`) on zero-length arrays              | `np.array([1,2,3]).mean()` → `2.0` |
| `std`, `var`     | Standard deviation and variance, respectively                               | `np.array([1,2,3]).std()` → `0.816...` |
| `min`, `max`     | Minimum and maximum                                                         | `np.array([1,2,3]).min()` → `1` |
| `argmin`, `argmax` | Indices of minimum and maximum elements, respectively                     | `np.array([1,2,3]).argmax()` → `2` |
| `cumsum`         | Cumulative sum of elements starting from 0                                  | `np.array([1,2,3]).cumsum()` → `[1,3,6]` |
| `cumprod`        | Cumulative product of elements starting from 1                              | `np.array([1,2,3]).cumprod()` → `[1,2,6]` |

## Array Set Operations (NumPy)

### Table 4-7. Array set operations with examples

| Method            | Description                                                                 | Example (Python) | Output |
|-------------------|-----------------------------------------------------------------------------|------------------|--------|
| `unique(x)`       | Compute the sorted, unique elements in `x`                                  | `np.unique([1,2,2,3,3,3])` | `[1 2 3]` |
| `intersect1d(x,y)`| Compute the sorted, common elements in `x` and `y`                          | `np.intersect1d([1,2,3],[2,3,4])` | `[2 3]` |
| `union1d(x,y)`    | Compute the sorted union of elements                                        | `np.union1d([1,2,3],[2,3,4])` | `[1 2 3 4]` |
| `in1d(x,y)`       | Compute a Boolean array indicating whether each element of `x` is in `y`    | `np.in1d([1,2,3],[2,4])` | `[False  True False]` |
| `setdiff1d(x,y)`  | Set difference: elements in `x` that are not in `y`                         | `np.setdiff1d([1,2,3,4],[2,3])` | `[1 4]` |
| `setxor1d(x,y)`   | Set symmetric differences: elements that are in either array, but not both  | `np.setxor1d([1,2,3],[2,3,4])` | `[1 4]` |

# 4.7 Example: Random Walks
